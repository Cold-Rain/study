Видим, что датасет "чистый": нет пропусков, количество, а все данные являются категориальными. Можно строить модель.

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

X = df[['Test', 'Age_Group']]
y = df['Status']

#преобразуем категории в нолики и единицы
encoder = OrdinalEncoder()
X_encoded = encoder.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=151)

#строим модель
model = CategoricalNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#проверяем качество модели
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
print(f'F1-Score: {f1:.4f}')
```

```python
import numpy as np

def f(x):
    return x**4 + 3*x**3 - 12*x**2 + 7*x - 2

def df(x):
    return 4*x**3 + 9*x**2 - 24*x + 7

def ddf(x):
    return 12*x**2 + 18*x - 24

#начальная точка
x = 10
tolerance = 0.0001 #точность

max_iterations = 10000 
iteration = 0

while iteration < max_iterations:
    derivative = df(x)
    second_derivative = ddf(x)
    
    if second_derivative == 0:
        print("Вторая производная равна 0.")
        break
    
    x_new = x - derivative / second_derivative
    change = abs(x_new - x)
    
    if change < tolerance:
        break
    
    x = x_new
    iteration += 1

if iteration < max_iterations:
    if ddf(x) > 0:
        print(f"Оптимальное значение x: {x:.4f}")
        print(f"Значение функции в точке х: {f(x):.4f}")
    else:
        print("Минимум функции не найден.")
```

Поскольку мы имеем дело с числом событий, произошедших за фиксированное время, то воспользуемся распределением Пуассона.
Поскольку за 8 часов рабочего дня в колл-центр поступает 16 звонков в среднем, то в час на коллстанцию в среднем поступает 2 звонка.
Соответственно 
$$
λ = 2 
$$
Построим распределение Пуассона (воспользуемся библиотеккой scipy и посчитаем сумму вероятностей для 2, 3 и 4 звонков

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
mu = 2
x = np.arange(0, 25, 1)
our_pmf = poisson.pmf(x, mu)
s = {}
for i in range(8):
    s[i] = round(our_pmf[i], 5)
print(s)
plt.scatter(x, our_pmf)
plt.title('PMF')
plt.grid() 
plt.show()
print(f'Ответ: вероятность, что в час поступит 2 до 4 звонков включительно = {s[2] + s[3] + s[4]:.3f}')
```

НО: в задании сказано, что пользоваться встроенными функциями построения распределения Пуассона нельзя нельзя, поэтому 
возьмем эту функцию
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

```python
import math

lambda_ = 2
probability = sum((lambda_**k * math.exp(-lambda_)) / math.factorial(k) for k in range(2, 5))
print(f'Ответ: вероятность, что в час поступит 2 до 4 звонков включительно = {probability:.3f}')
```

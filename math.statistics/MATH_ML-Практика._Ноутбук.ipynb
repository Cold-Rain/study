Выведите на экран полученные оценки коэффициентов модели, округлённые до целого.

````python
#Поскольку матрица наблюдений и вектор правильных ответов уже созданы
#приступим к созданию модели линейной регрессии по методу наименьших квадратов
# Добавляем столбец единиц для свободного члена (intercept)
X = np.hstack([np.ones((X.shape[0], 1)), X])

#Создаем матрицу Грама
XTX_inv = np.linalg.inv(X.T @ X)
#Вычисляем коэффициенты модели
beta_hat = XTX_inv @ X.T @ y

# Округление коэффициентов до целых
beta_hat_rounded = np.round(beta_hat).astype(int)

# Вывод коэффициентов
print("Оценки коэффициентов модели (округленные до целых):")
print(beta_hat_rounded)
````

**Выводы:**<br>

Признак Brittle имеет коэффициент 25, что относительно мало по сравнению с другими коэффициентами. Это может указывать на то, что хрупкость оказывает меньшее влияние на продуктивность по сравнению с другими признаками.<br>
AI и TOC оказывают отрицательное влияние на продуктивность, что может указывать на то, что более высокие значения акустического импеданса и общего органического углерода снижают продуктивность.<br>
VR оказывает значительное положительное влияние на продуктивность, что может указывать на то, что более высокая скорость распространения волны связана с более высокой продуктивностью.<br>

````python
# переписываем словарик с параметрами скважины
well_params = {
    'Well': 106.0,
    'Por': 15.32,
    'Perm': 3.71,
    'AI': 3.29,
    'Brittle': 55.99,
    'TOC': 1.35,
    'VR': 2.42
}

# Извлекаем признаки для прогноза
x_pred = np.array([1, well_params['Por'], well_params['Perm'],
                   well_params['AI'], well_params['Brittle'],
                   well_params['TOC'], well_params['VR']])

# Вычисляем прогнозируемое значение
y_pred = np.dot(x_pred, beta_hat)

# Известное значение выработки газа
y_true = 4748.315024

# Рассчитываем абсолютную ошибку
absolute_error = np.abs(y_true - y_pred)

print(f"Прогнозируемое значение выработки газа: {y_pred:.2f}")
print(f"Абсолютная ошибка: {absolute_error:.2f}")
````
